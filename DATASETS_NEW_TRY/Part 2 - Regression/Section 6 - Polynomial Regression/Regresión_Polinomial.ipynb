# <span style = 'color:green'> Regresión Polinomial

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Librerías para categorizar las variables
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, PolynomialFeatures
from sklearn.compose import ColumnTransformer

#Librerias para la division de los conjuntos de entrenamiento
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
```

```python
ds = pd.read_csv("Position_Salaries.csv")
```

# <span style = 'color:red'>  1. Separación de variables

```python
x = ds.iloc[:,1:2].values
y = ds.iloc[:,2].values
```

# <span style = color:red> 2. Ajuste de regresión lineal y polinómica 

```python
#Regresión lineal
lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

```python
#Regresión polinomial
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(x)

#Se realiza la misma clase para la regresión polinomial
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly,y)
```

# <span style = 'color:red'> 3. Visualización de los modelos

```python
#Modelo lineal
plt.scatter(x,y, color = 'red')
plt.plot(x,lin_reg.predict(x), color = 'blue')
plt.title("Regresión Lineal")
plt.xlabel("Niveles")
plt.ylabel("Salario")
plt.show()
```

```python
#Modelo polinómico
plt.scatter(x,y, color = 'red')
plt.plot(x,lin_reg_2.predict(x_poly), color = 'blue')
plt.title("Regresión Polinómica")
plt.xlabel("Niveles")
plt.ylabel("Salario")
plt.show()
```
